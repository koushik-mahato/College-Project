Importing the Dependencies

```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

Data Collection and Processing

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
# Loading data from sklearn
print (breast_cancer_dataset)
```

```python
# Loading datato a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# Print the dataset
data_frame.head()
```

```python
# Adding the 'target' columnm to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
# print the last 5 rows of the dataframe
data_frame.tail()
```

```python
# number of rows and columns in the dataset
data_frame.shape
```

```python
# getting some info aboutthe data
data_frame.info()
```

```python
# checking for missing values
data_frame.isnull().sum()
```

```python
# statistical measures about the data
data_frame.describe()
```

```python
# checking the distribution of target variable
data_frame['label'].value_counts()
```

0 --> Malignant

1 --> Benign

```python
data_frame.groupby('label').mean()
```

Separate the features and target

```python
X = data_frame.drop(columns = 'label', axis = 1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape,X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the logistic regression model using training data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)*100
```

```python
print('Accuracy on training data = ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)*100
```

```python
print('Accuracy on test data = ',test_data_accuracy)
```

```python
# Classification report with precision, recall, and F1-score
print("Classification Report:")
print(classification_report(Y_test, X_test_prediction))
```

Building a Predictive System

```python
input_data = (17.99,10.38,122.8,1001,0.1184,0.2776,0.3001,0.1471,0.2419,0.07871,1.095,0.9053,8.589,153.4,0.006399,0.04904,0.05373,0.01587,0.03003,0.006193,25.38,17.33,184.6,2019,0.1622,0.6656,0.7119,0.2654,0.4601,0.1189)

# change the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array as we are predicting for one datapoint
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if(prediction[0] == 0):
  print('The Breast Cancer is Malignent')
else:
  print('The Breast Cancer is Benign')
```
